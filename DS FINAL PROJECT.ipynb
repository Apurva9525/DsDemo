About Dataset Internet Speed Prediction Dataset

Ping_latency: Time taken for a signal to travel from the source to the destination.

Download_speed: Rate at which data is transferred from the internet to the user's device.

Upload_speed: Rate at which data is transferred from the user's device to the internet.

Packet_loss_rate: Proportion of data packets lost during transmission.

Router_distance: Physical distance between the user's device and the router.

Network_congestion: Level of congestion within the network.

ISP_quality: Quality of service provided by the Internet Service Provider (ISP).

Connection_type_DSL, Connection_type_Cable, Connection_type_Fiber: Binary indicators for the type of internet connection.

Signal_strength: Strength of the wireless signal received by the user's device.

Weather_conditions: Environmental factors that may affect internet performance.





Target Feature: Stability_Category

```python
max_speed = df['Internet_speed'].max()  
bins = [0, 500, 1500, 2500, max_speed + 1]  
labels = ["Very Unstable", "Unstable", "Stable", "Very Stable"]

df['Stability_Category'] = pd.cut(df['Internet_speed'], bins=bins, labels=labels)     #cut() : used for the equal binning
```

```python
df['Stability_Category'].replace(["Very Unstable", "Unstable", "Stable", "Very Stable"],[1,2,3,4],inplace=True)
```

```python
x= df[["Ping_latency", "Download_speed", "Upload_speed", "Packet_loss_rate","Network_congestion", "Connection_type_DSL", "Connection_type_Cable", 
        "Connection_type_Fiber", "Signal_strength", "Weather_conditions"]]
y=df["Stability_Category"]
```
